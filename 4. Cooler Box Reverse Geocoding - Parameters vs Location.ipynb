```python
import pandas as pd
import numpy as np
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import tqdm
from tqdm.notebook import tqdm_notebook
```

# 4.1 Reverse Geocoding to Extract Address of Temp with Location

```python
Temp=pd.read_excel('Temp with Location.xlsx')
Temp.head()
```

```python
Temp['Geom'] = Temp['lat'].map(str) + ',' + Temp['long'].map(str)
Temp.head()
```

```python
locator = Nominatim(user_agent='myGeocoder', timeout=10)
rgeocode = RateLimiter(locator.reverse, min_delay_seconds=0.001)
```

```python
tqdm_notebook.pandas()
Temp['Address'] = Temp['Geom'].progress_apply(rgeocode)
Temp.head()
```

```python
Temp.to_excel('Address of Temp with Location.xlsx',index=False)
```
